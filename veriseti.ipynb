```python
# Cinsiyete göre kalp rahatsızlığı dağılımını hesapla
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("heart_disease.csv")

gender_counts = df.groupby(['sex', 'Disease'])['Disease'].count().unstack()

# Cinsiyet etiketlerini oluştur
gender_labels = {0: "Erkek", 1: "Kadın"}
gender_counts = gender_counts.rename(index=gender_labels)


# Grafiği çiz
gender_counts.plot(kind='bar', stacked=True)
plt.title("Cinsiyete Göre Kalp Rahatsızlığı Dağılımı")
plt.xlabel("Cinsiyet")
plt.ylabel("Hasta Sayısı")
plt.legend(["Rahatsızlık Yok", "Rahatsızlık Var"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("heart_disease.csv")

# Kalp rahatsızlığı olan ve olmayanların yaş dağılımını hesapla
disease_counts = df.groupby(['age', 'Disease'])['Disease'].count().unstack()

# Grafiği çiz
disease_counts.plot(kind='bar', stacked=True)
plt.title("Yaşlara Göre Kalp Rahatsızlığı Dağılımı")
plt.xlabel("Yaş")
plt.ylabel("Hasta Sayısı")
plt.legend(["Rahatsızlık Yok", "Rahatsızlık Var"])
plt.show()
```

```python
# Şeker hastalığına göre kalp rahatsızlığı dağılımını hesapla
diabetes_counts = df.groupby(['fasting blood sugar > 120 mg/dl ', 'Disease'])['Disease'].count().unstack()

# Şeker hastalığı etiketlerini güncelle
diabetes_labels = {0: "Şeker Hastalığı Yok", 1: "Şeker Hastalığı Var"}
diabetes_counts = diabetes_counts.rename(index=diabetes_labels)

# Grafiği çiz
diabetes_counts.plot(kind='bar', stacked=True)
plt.title("Şeker Hastalığına Göre Kalp Rahatsızlığı Dağılımı")
plt.xlabel("Şeker Hastalığı")
plt.ylabel("Hasta Sayısı")
plt.legend(["Rahatsızlık Yok", "Rahatsızlık Var"])
plt.show()
```
